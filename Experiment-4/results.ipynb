```python
import pandas as pd
import numpy as np
import json
```

```python

models = [
    'Gradient Boosting',
    'K-Nearest Neighbors',

    'XGBClassifier',

    'Logistic Regression',
    'Ridge',
    'Extra Trees',
    'AdaBoost',

    'Passive Aggressive',
    'Support Vector Classification',
    'Decision Trees',

    'Random Forest',
    'Naive Bayes'

]
```

```python
len(models)
```

```python
result = {}
```

```python
for model in models:
    path = model + '.json'
    with open(path, 'r') as f:
        data = json.load(f)
    result[model] = [data['Train Accuracy'], data['Test Accuracy'], data['Train F1'], data['Test F1'], data['Best Score'], data['Best Params']]
```

```python
results = pd.DataFrame(result).T
results.columns = ['Train Accuracy', 'Test Accuracy', 'Train F1', 'Test F1', 'Best Score', 'Best Params']
results
```
