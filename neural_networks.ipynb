```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


import numpy as np
```

Load the data from the uploaded files

```python

data = pd.read_csv("Data\\BP_features.csv")
labels = pd.read_csv("Data\\final_labels.csv")
labels = labels.iloc[:, 1]
```

```python
count_normal = 0
for i in labels:
    count_normal += i
count_normal
```

Splitting the datasets into training and testing sets

```python

# Split ratio = 80:20 -> (441: 111)
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42, shuffle=True)
```

```python
count_normal = 0
for i in y_train:
    count_normal += i
count_normal
```

Scaling the data

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

Train and evaluate model

```python
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

# Function to create model, required for KerasClassifier
def create_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python

# Create a model using KerasClassifier
model = KerasClassifier(build_fn=create_model, epochs=1000, batch_size=32)
```

```python
history = model.fit(X_train, y_train)
```

```python
# Evaluate the model on the test set
model.score(X_test, y_test)
# print(f"Test Accuracy: {accuracy*100:.2f}%")
```

```python
y_pred = model.predict(X_test)
```
